```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
full = pd.read_csv('IL-clean.csv',low_memory=False)
```

```python
full.head()
```

```python
full.columns
```

```python
def SummaryStats(df,label):
    n_stops = df.shape[0]
    n_searches = df['search_conducted'].sum()
    n_hits = df['contraband_found'].sum()
    search_rate = n_searches / n_stops
    hit_rate = n_hits / n_searches
    td={'n_stops' : n_stops, 'n_searches' : n_searches, 'n_hits' : n_hits, 'search_rate' : search_rate,'hit_rate' : hit_rate}
    return pd.DataFrame(data=td,index=[label])
```

```python
temptimes = full['stop_time'][:].str.slice(0,2)
times = pd.to_numeric(temptimes)
```

```python
hit_rate = []
n_hits = []
n_searches = []
n_stops = []
search_rate = []
for i in range(0,24):
    temp = SummaryStats(full[times==i],'temp')
    hit_rate.append(temp.get_value('temp','hit_rate'))
    n_hits.append(temp.get_value('temp','n_hits'))
    n_searches.append(temp.get_value('temp','n_searches'))
    n_stops.append(temp.get_value('temp','n_stops'))
    search_rate.append(temp.get_value('temp','search_rate'))
    
```

```python
fig, ax1 = plt.subplots()
ax1.plot(range(0,24), n_stops, 'b-')
ax1.set_xlabel('Hour of Day')
ax1.set_ylabel('Stops', color='b')
ax1.tick_params('y', colors='b')
ax2 = ax1.twinx()
ax2.plot(range(0,24), n_searches, 'r-')
ax2.set_ylabel('Searches', color='r')
ax2.tick_params('y', colors='r')
```

```python
fig, ax1 = plt.subplots()
ax1.plot(range(0,24), search_rate, 'b-',label='Search Rate')
ax1.set_ylabel('Search Rate', color='b')
ax1.tick_params('y', colors='b')
plt.xlabel('Hour of Day')
ax2=ax1.twinx()
ax2.plot(range(0,24), hit_rate, 'r-',label='Hit Rate')
ax2.set_ylabel('Hit Rate', color='r')
ax2.tick_params('y', colors='r')
```

```python
ages = pd.to_numeric(full['driver_age'])
m_hit_rate = []
m_n_hits = []
m_n_searches = []
m_n_stops = []
m_search_rate = []
f_hit_rate = []
f_n_hits = []
f_n_searches = []
f_n_stops = []
f_search_rate = []
for i in range(16,100):
    temp = SummaryStats(full[(ages==i) & (full['driver_gender']=='M')],'temp')
    m_hit_rate.append(temp.get_value('temp','hit_rate'))
    m_n_hits.append(temp.get_value('temp','n_hits'))
    m_n_searches.append(temp.get_value('temp','n_searches'))
    m_n_stops.append(temp.get_value('temp','n_stops'))
    m_search_rate.append(temp.get_value('temp','search_rate'))
    temp = SummaryStats(full[(ages==i) & (full['driver_gender']=='F')],'temp')
    f_hit_rate.append(temp.get_value('temp','hit_rate'))
    f_n_hits.append(temp.get_value('temp','n_hits'))
    f_n_searches.append(temp.get_value('temp','n_searches'))
    f_n_stops.append(temp.get_value('temp','n_stops'))
    f_search_rate.append(temp.get_value('temp','search_rate'))
```

```python
fig, ax = plt.subplots()
ax.plot(range(16,100), m_n_stops, 'b-',label='Male Stops')
ax.plot(range(16,100), f_n_stops, 'm-',label='Female Stops')
ax.set_ylabel('Total Stops')
ax.tick_params('y')
plt.xlabel('Age')
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(range(16,100), m_n_stops/max(m_n_stops), 'b-',label='Male Stops')
ax.plot(range(16,100), f_n_stops/max(f_n_stops), 'm-',label='Female Stops')
ax.set_ylabel('Normalized Stops')
ax.tick_params('y')
plt.xlabel('Age')
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(range(16,100), m_n_searches, 'b-',label='Male Searches')
ax.plot(range(16,100), f_n_searches, 'm-',label='Female Searches')
ax.set_ylabel('Total Searches')
ax.tick_params('y')
plt.xlabel('Age')
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(range(16,100), m_search_rate, 'b-',label='Male Search Rate')
ax.plot(range(16,100), f_search_rate, 'm-',label='Female Search Rate')
ax.set_ylabel('Total Searches')
ax.tick_params('y')
plt.xlabel('Age')
plt.legend()
```

```python
fig, ax1 = plt.subplots()
ax1.plot(range(16,100), m_search_rate, 'b-',label='Search Rate')
ax1.set_ylabel('Male Search Rate', color='b')
ax1.tick_params('y', colors='b')
plt.xlabel('Age')
ax2=ax1.twinx()
ax2.plot(range(16,100), m_hit_rate, 'r-',label='Hit Rate')
ax2.set_ylabel('Male Hit Rate', color='r')
ax2.tick_params('y', colors='r')
```

```python
fig, ax1 = plt.subplots()
ax1.plot(range(16,100), f_search_rate, 'b-',label='Search Rate')
ax1.set_ylabel('Female Search Rate', color='b')
ax1.tick_params('y', colors='b')
plt.xlabel('Age')
ax2=ax1.twinx()
ax2.plot(range(16,100), f_hit_rate, 'r-',label='Hit Rate')
ax2.set_ylabel('Female Hit Rate', color='r')
ax2.tick_params('y', colors='r')
```

```python
SummarizedStops = SummaryStats(full,'AllStops')
for race in full['driver_race'].unique():
    temp = SummaryStats(full[full['driver_race']==race], race+'Stops')
    SummarizedStops = SummarizedStops.append(temp)    

SummarizedStops
```

```python
LocationStops = SummaryStats(full,'Total')
LocationStops['race'] = 'All'
LocationStops['location'] = 'All'
LocationStops['White_hit_rate'] = SummarizedStops.loc['WhiteStops']['hit_rate']
LocationStops['White_search_rate'] = SummarizedStops.loc['WhiteStops']['search_rate']
for district in full['district'].unique():
    if not isinstance(district,str):
        continue
    tempwhite = SummaryStats(full[(full['driver_race']=='White') & (full['district']==district)], 'White' + ' in ' + district)
    whitehit = tempwhite.get_value('White' + ' in ' + district,'hit_rate')
    whitesearch = tempwhite.get_value('White' + ' in ' + district,'search_rate')
    for race in full['driver_race'].unique():
        #print( race + ' in ' + district)        
        if race=='White':
            temp = tempwhite
        elif race=='Other':
            continue
        else: 
            temp = SummaryStats(full[(full['driver_race']==race) & (full['district']==district)], race + ' in ' + district)
        temp['race'] = race
        temp['location'] = district
        temp['White_hit_rate'] = whitehit
        temp['White_search_rate'] = whitesearch
        LocationStops = LocationStops.append(temp)
```

```python
ByRace = LocationStops.groupby('race')
```

```python
fig, ax = plt.subplots()
for name, group in ByRace:
    ax.plot(group.White_search_rate, group.search_rate, marker='o',linestyle='',label=name)
    
ax.legend()
plt.xlabel('White Search Rate in District')
plt.ylabel('Group Search Rate')
```

```python
fig, ax = plt.subplots()
for name, group in ByRace:
    ax.plot(group.White_hit_rate, group.hit_rate, marker='o',linestyle='',label=name)
    
ax.legend()
plt.xlabel('White Hit Rate in District')
plt.ylabel('Group Hit Rate')
```

```python
LocationStops['diff_hit_rate']=LocationStops['hit_rate']/LocationStops['White_hit_rate']
LocationStops['diff_search_rate']=LocationStops['search_rate']/LocationStops['White_search_rate']
```

```python
fig, ax = plt.subplots()
for name, group in ByRace:
    ax.plot(group.diff_search_rate, group.diff_hit_rate, marker='o',linestyle='',label=name)
    
ax.legend()
plt.xlabel('Relative Search Rate')
plt.ylabel('Relative Hit Rate')
```
